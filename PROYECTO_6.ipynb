```python
!pip install squarify
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
import plotly
import squarify as sq
from IPython.core.pylabtools import figsize
```

##Exportaciones

```python
Country_export=pd.read_excel("/content/Trade_Map_-_List_of_importing_markets_for_a_product_exported_by_Colombia.xlsx")
```

```python
Country_export.columns
```

```python
Country_export.columns=['Importers', '2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
```

```python
Filter_DataEx=Country_export.iloc[:6]
```

```python
Filter_DataEx
```

```python
DataEx=Filter_DataEx.set_index('Importers')
```

```python
DataEx
```

```python
plt.barh(Filter_DataEx['Importers'], Filter_DataEx['2018'], color='blue')
plt.ylabel('Paises')
plt.xlabel('Millones de USD')
plt.title('Exportaciones 2018')
```

```python
plt.barh(Filter_DataEx['Importers'], Filter_DataEx['2019'], color='yellow')
plt.ylabel('Paises')
plt.xlabel('Millones de USD')
plt.title('Exportaciones 2019')
```

```python
plt.barh(Filter_DataEx['Importers'], Filter_DataEx['2020'], color='green')
plt.ylabel('Paises')
plt.xlabel('Millones de USD')
plt.title('Exportaciones 2020')
```

```python
for i in range(2018,2021):

  fig, ax = plt.subplots(figsize=(12,8))

  sizes = DataEx[str (i)].iloc[1:6].values
  labels = DataEx[str (i)].iloc[1:6].index
  colors = sns.color_palette('flare')
  sq.plot(sizes=sizes, label=labels, text_kwargs=dict(color='white'))
  plt.title(i)
  plt.axis('off')
  plt.show()
```

```python
DataEx.loc['World'].plot()
plt.xlabel('Años')
plt.ylabel('Cantidad de importaciones en USD')
plt.title('Exportaciones en el tiempo del mundo')
```

```python
DF1=DataEx.transpose()
```

```python
DF1.plot(figsize(10,5), xlabel='Año', ylabel='Cantidad de importaciones en USD', title='Exportaciones en el tiempo para Colombia')
```
